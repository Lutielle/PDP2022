```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from pandas import read_csv
from pandas.plotting import scatter_matrix
from matplotlib import pyplot

dataset = read_csv("C:\\Users\\Kai\\Documents\\Work\\ES5m-sma.txt")
pd.options.display.max_rows = 4000
#dataset.head(20)
columnname = ['sma200', 'stoch100', 'rsi60', 'momentum20']
X = dataset[columnname]
#X

y = dataset['direction']
#y


dataset
```

```python
from sklearn.preprocessing import StandardScaler
standardizer = StandardScaler()
X = standardizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.25, shuffle=False)
#random_state = 0 renders the randomization between iterations consistent, allowing for debugging/testing/detecting changes between runs
#shuffle = False causes the data to be selected in order for use for testing, instead of random data points in the set (True is the default state)
#setting shuffle = False decreases the accuracy of the prediction significantly, but using randomly ordered data generates outright incorrect values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train) #fit() is the training method
```

```python
predictions = model.predict(X_test) #generate predictions using trained model
predictions
```

```python
#predictions = model.predict(X_test) #generate predictions using trained model

#print(predictions.size)
#print(dataset['close'].size)
#print(y_train.size)

#predictions = predictions; predP = predictions and direction both positive;
#N = predictions and direction both negative; W = predictions and direction do not match

#initializes columns
#then sets initial segment (where there is training data) to 0
dataset['predictions'] = dataset['close']
dataset['predP'] = dataset['close']
dataset['predN'] = dataset['close']
dataset['predW'] = dataset['close']

for i in range(y_train.size):
    dataset['predictions'][i] = 0
    dataset['predP'][i] = 0
    dataset['predN'][i] = 0
    dataset['predW'][i] = 0

#print(predictions)
dataset
```

```python
#assigns prediction column, for error checking
count = 0
for i in range(y_train.size, dataset['predictions'].size):
    dataset['predictions'][i] = predictions[count]       
    count += 1
```

```python
#assigns predP, N, W
count = 0
for i in range(y_train.size, dataset['predP'].size):
    if dataset['direction'][i] == 1 and predictions[count] == 1:
        dataset['predP'][i] = dataset['close'][i]
        dataset['predN'][i] = 0
        dataset['predW'][i] = 0
    elif dataset['direction'][i] == 0 and predictions[count] == 0:
        dataset['predP'][i] = 0
        dataset['predN'][i] = dataset['close'][i]
        dataset['predW'][i] = 0    
    else:
        dataset['predP'][i] = 0
        dataset['predN'][i] = 0
        dataset['predW'][i] = dataset['close'][i]
    count += 1

#assigns predB
dataset['predB'] = dataset['predP'] + dataset['predN']
```

```python
#assigns predP2, N2, W2 - previous columns but with offset for future change margin
dataset['predP2'] = dataset['close']
dataset['predN2'] = dataset['close']
dataset['predW2'] = dataset['close']

count = 0
offset = 1
for i in range(y_train.size, dataset['predP'].size - offset):
    if dataset['direction'][i] == 1 and predictions[count] == 1:
        dataset['predP2'][i + offset] = dataset['close'][i]
        dataset['predN2'][i + offset] = 0
        dataset['predW2'][i + offset] = 0
    elif dataset['direction'][i] == 0 and predictions[count] == 0:
        dataset['predP2'][i + offset] = 0
        dataset['predN2'][i + offset] = dataset['close'][i]
        dataset['predW2'][i + offset] = 0    
    else:
        dataset['predP2'][i + offset] = 0
        dataset['predN2'][i + offset] = 0
        dataset['predW2'][i + offset] = dataset['close'][i]
    count += 1
    
#assigns predB2
dataset['predB2'] = dataset['predP2'] + dataset['predN2']

dataset.tail(500)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predictions)

TN, FP, FN, TP = confusion_matrix(y_test, predictions).ravel()

print('True Positive(TP)  = ', TP)
print('False Positive(FP) = ', FP)
print('True Negative(TN)  = ', TN)
print('False Negative(FN) = ', FN)

accuracy =  (TP+TN) /(TP+FP+TN+FN)

print('Accuracy of the binary classification = {:0.3f}'.format(accuracy))
```

```python
models = {}

# Logistic Regression
from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

# Support Vector Machines
from sklearn.svm import LinearSVC
models['Support Vector Machines'] = LinearSVC()

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
models['K-Nearest Neighbor'] = KNeighborsClassifier()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy, precision, recall = {}, {}, {}

for key in models.keys():
    
    # Fit the classifier model
    models[key].fit(X_train, y_train)
    
    # Prediction 
    predictions = models[key].predict(X_test)
    
    # Calculate Accuracy, Precision and Recall Metrics
    accuracy[key] = accuracy_score(predictions, y_test)
    precision[key] = precision_score(predictions, y_test)
    recall[key] = recall_score(predictions, y_test)
```

```python
df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df_model['Accuracy'] = accuracy.values()
df_model['Precision'] = precision.values()
df_model['Recall'] = recall.values()

df_model
```
